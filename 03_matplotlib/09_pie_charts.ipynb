```python
#---Pie----charts--------------
import matplotlib.pyplot as plt
import numpy as np

x = np.array([36, 25, 25, 15])
#--using---labels----------
mylabels = ["apple", "Banana", "chery", "orange"]
plt.pie(x, labels =mylabels)
plt.show()
```

```python
#---------startangle-------------------
import matplotlib.pyplot as plt
import numpy as np

x = np.array([36, 25, 25, 15])
#--using---labels----------
mylabels = ["apple", "Banana", "chery", "orange"]


plt.pie(x, labels =mylabels, startangle=270)
plt.show()
```

```python
#---------explod----shadow---------------
import matplotlib.pyplot as plt
import numpy as np

x = np.array([36, 25, 25, 15])
#--using---labels----------
mylabels = ["apple", "Banana", "chery", "orange"]

#---explod----------
myexplod = [0.2, 0.1, 0, 0]
plt.pie(x, labels =mylabels, startangle=270 , explode=myexplod, shadow=True)
plt.show()
```

```python
#----colors--legend--width--header-----------
import matplotlib.pyplot as plt
import numpy as np

x = np.array([36, 25, 25, 15])
#--using---labels----------
mylabels = ["apple", "Banana", "chery", "orange"]

#---explod----------
myexplod = [0.2, 0.1, 0, 0]

#--colors
mycolor = ["#4CAF50", "hotpink", "y", "b"]
plt.pie(x, labels =mylabels, startangle=270 , explode=myexplod, shadow=True ,colors=mycolor )

#---legend-----
plt.legend(title="fruits")
plt.show()
```

```python
#----------%f------------
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# Show percentage values on the pie chart
plt.pie(y, labels=mylabels, autopct='%1.1f%%')

plt.legend(title="Fruits")
plt.show()
```
