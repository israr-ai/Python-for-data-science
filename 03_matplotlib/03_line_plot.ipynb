```python
#---line---plotting---
import matplotlib.pyplot as plt
import numpy as np
y = np.array([10, 20, 25, 30])
plt.plot(y ,linestyle='dotted')
plt.show()
```

```python
#---line--color--and--width----
import matplotlib.pyplot as plt
import numpy as np
y = np.array([10, 20, 25, 30])
# plt.plot(y, color='green', linewidth=5)
plt.plot(y, c='green', lw=5)
plt.show()
```

```python
#----multiple--points--with--lines--------
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
#---Draw two lines by specifiyng the x- and y-point values for both lines:
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
