```python
#--pandas---plotting- lineplot-------------
import pandas as pd
import matplotlib.pyplot as plt

#sample data----
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "sales" : [150, 300, 450, 600, 750],
    "profit": [50, 100, 150, 200, 250]
}
df = pd.DataFrame(data)

#-plot all--numeric---values
df.plot(x="Year", y=["sales", "profit"], kind="line", marker="o")
plt.title("Sales and Profit Over Years")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.grid(True)
plt.legend(title="Metrics")
plt.show()
```

```python
#----Bar--plot----------------
import pandas as pd
import matplotlib.pyplot as plt

#sample data----
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "sales" : [150, 300, 450, 600, 750],
    "profit": [50, 100, 150, 200, 250]
}
df = pd.DataFrame(data)
#-plot all--numeric---values
df.plot(x="Year", y=["sales", "profit"], kind="bar")
plt.title("sales and Profit Over Years")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.grid(axis="y")
plt.legend(title="Metrics")
plt.show()
```

```python
#---Histogram---plot----------------
import pandas as pd
import matplotlib.pyplot as plt

#sample data----
data ={
    "Year": [2018, 2019, 2020, 2021, 2022],
    "sales" : [150, 300, 450, 600, 750],
    "profit": [50, 100, 150, 200, 250]
}
df = pd.DataFrame(data)
#-plot all--numeric---values
df[["sales", "profit"]].plot(kind="hist", bins=5, alpha=0.7)
plt.title("Distribution of Sales and Profit")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.grid(axis="y")
plt.legend(title="Metrics")
plt.show()
```

```python
#---Box--plot----------------
import pandas as pd
import matplotlib.pyplot as plt

#sample data----
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "sales" : [150, 300, 450, 600, 750],
    "profit": [50, 100, 150, 200, 250]
}
df = pd.DataFrame(data)
#-plot all--numeric---values
df[["sales", "profit"]].plot(kind="box")
plt.title("Box Plot of Sales and Profit")
plt.ylabel("Amount")
plt.grid(axis="y")
plt.show()
```

```python
#--Scatter--plot----------------
import pandas as pd
import matplotlib.pyplot as plt

#sample data----
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "sales" : [150, 300, 450, 600, 750],
    "profit": [50, 100, 150, 200, 250]
}
df = pd.DataFrame(data)
#-plot all--numeric---values
df.plot(x="sales", y="profit", kind="scatter", color="blue", s=100)
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
#---correlation---heatmap--(extra with seaborn)----------------
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#sample data----
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "sales" : [150, 300, 450, 600, 750],
    "profit": [50, 100, 150, 200, 250]
}
df = pd.DataFrame(data)
#-compute correlation matrix
corr = df.corr()
#-plot heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
